### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (top-left)
x = np.arange(0, 11)
y = x ** 2
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y = X^2')

# Bar chart (top-right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')

# Scatter plot (bottom-left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
for xi, yi in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({xi},{yi})', (xi, yi), textcoords="offset points", xytext=(5,5), ha='left')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Pie chart (bottom-right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

plt.tight_layout()
plt.show()
```
